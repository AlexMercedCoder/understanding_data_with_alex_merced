```python
import polars as pl
import seaborn as sns

# Creating the DataFrame
data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05",
             "2023-01-06", "2023-01-07", "2023-01-08", "2023-01-09", "2023-01-10",
             "2023-01-11", "2023-01-12", "2023-01-13", "2023-01-14", "2023-01-15",
             "2023-01-16", "2023-01-17", "2023-01-18", "2023-01-19", "2023-01-20"],
    "Product": ["A", "B", "A", "C", "B", "C", "A", "B", "C", "A",
                "B", "A", "C", "B", "A", "C", "B", "A", "C", "B"],
    "Sales": [200, 150, 180, 220, 170, 210, 190, 160, 230, 205,
              165, 175, 215, 155, 195, 225, 145, 185, 235, 150]
}

df = pl.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
# lineplot of sales by date
sns.lineplot(data=df, x="Date", y="Sales").set(title="Sales by Date")
```

```python
# barplot of sales by product
sns.barplot(data=df, x="Product", y="Sales").set(title="Sale by Product")
```
